```python
import ctypes
import time
import os
import subprocess
import matplotlib.pyplot as plt

_libInc = ctypes.CDLL('./myLib/libMyLib.so')
```

```python
def cpu_select(cpu_core, fib_term):
    isValid = False
    if cpu_core == 0 or cpu_core == 1:
        isValid = True
    else:
        print("Incorrect core option")

    if isValid:
        try:
            #os.system(f"taskset -c {user_in} python3 fib.py")
            cp = subprocess.run(["taskset", "-c", f"{cpu_core}", "python3", "fib.py"], capture_output=True, input=f"{fib_term}\n", text=True)
            result = cp.stdout.strip()
            result_d = float(result)
        except Exception as e:
            print(e)
    return result_d
```

```python
import math
def plot_cycles():
    fib_terms = [5,10,25,30]
    num_trials = 3
    x_terms = []
    y1_count = []
    y2_time = []
    error_count = []
    error_time = []
    cpu_freq = 650000000
    for term in fib_terms:
        trial_cycle_count = []
        trial_cycle_time = []
        mean_count = 0
        mean_time = 0
        
        x_terms.append(term)
        for trial in range(num_trials):
            _libInc.inititialize(1,0)
            before_cycle_count = _libInc.cycle_count()
            # run on cpu 1
            # execution time is calculated by fib.py
            before_t = time.time()
            cpu_select(1, term)
            after_cycle_count = _libInc.cycle_count()
            after_t = time.time()
            exec_time = after_t - before_t
            cycle_count = after_cycle_count - before_cycle_count
            trial_cycle_time.append(exec_time)
            trial_cycle_count.append(cycle_count/cpu_freq) # dividing by cpu freq will give us the seconds of execution
            
        mean_time = sum(trial_cycle_time)/num_trials
        mean_count = sum(trial_cycle_count)/num_trials
        print("Total cycles: ")
        print(mean_count)
        print("Total Execution Time: ")
        print(mean_time)
        y1_count.append(mean_count)
        y2_time.append(mean_time)
        
        trial_cycle_count = [(count-mean_count)**2 for count in trial_cycle_count]
        sd_count = math.sqrt((1/(num_trials-1))*sum(trial_cycle_count))
        trial_cycle_time = [(time-mean_time)**2 for time in trial_cycle_time]
        sd_time = math.sqrt((1/(num_trials-1))*sum(trial_cycle_time))

        error_count.append(sd_count) 
        error_time.append(sd_time) 
    #plotting figure with Cycles vs fib terms
    plt.figure()
    plt.errorbar(x_terms, y1_count, yerr=error_count, fmt='o', capsize=4)
    plt.xlabel("fib terms(n)")
    plt.ylabel("Execution Time(s)")
    plt.title("PMU Counter")
    plt.grid(True)
    plt.show()
    # Plotting figure with execution time vs fib terms
    plt.figure()
    plt.errorbar(x_terms, y2_time, yerr=error_time, fmt='o', capsize=4)
    plt.xlabel("fib terms(n)")
    plt.ylabel("Execution Time(s)")
    plt.title("Timing Module")
    plt.grid(True)
    plt.show()
        
        
```

```python
plot_cycles()
```
