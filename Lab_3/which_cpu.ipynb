```python
import os
import subprocess

isValid = False

user_in=input("Enter core to use: ")

if user_in == "0" or user_in == "1":
    isValid = True
else:
    print("Incorrect core option")
    
if isValid:
    try:
        #os.system(f"taskset -c {user_in} python3 fib.py")
        subprocess.run(["taskset", "-c", f"{user_in}", "python3", "fib.py"], capture_output=False, input="35\n", text=True)
    except Exception as e:
        print(e)
```
